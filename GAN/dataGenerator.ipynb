### Data generation

```python
from keras.models import Sequential, load_model
from numpy import expand_dims, ones, zeros, asarray
from numpy.random import randn
from numpy.random import rand
from numpy.random import randint
import pickle
from matplotlib import pyplot
import numpy as np
```

```python
latent_dim = 100
dataSize = 10000
sequenceLen = 95
dimensionsCount = 6
```

```python
model = load_model('bestGenerator/generator_model_2000.h5')
vector = asarray([rand(dataSize, latent_dim)])
vector = vector.reshape(dataSize,latent_dim)
# generate data
X = model.predict(vector)
X = np.reshape(X,(dataSize,sequenceLen,dimensionsCount),order='c')
pickle.dump(X, open('generatedData/3/generatedData.pickle', 'wb'))
```

```python


with open('generatedData/3/generatedData.pickle', 'rb') as data:
    training_set = pickle.load(data)
    
training_set.shape

fig, axes= pyplot.subplots(nrows=6, ncols=1,figsize=(10,20))
for i in range(6):
    # plot in subplots
    axes[i].plot(training_set[10, :, i])#sequenceLen*dimensionsCount,1
    
pyplot.show()
```
